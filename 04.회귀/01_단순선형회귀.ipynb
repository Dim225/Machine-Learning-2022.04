# 단순 선형회귀

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 아버지와 아들의 키 - Galton

```python
df = pd.read_csv('http://www.randomservices.org/random/data/Galton.txt', sep='\t')
df.head()
```

- 아버지와 아들의 키만 발췌
- 컬럼명을 Father, Son 으로 변경
- 단위 변경 (inch => centi)

```python
# Filtering : 아들의 키
df = df[df.Gender == 'M']
df.head(3)
```

```python
# Selection
df = df[['Father', 'Height']]
df.head(3)
```

```python
# 칼럼명 변경
df.rename(columns={'Height':'Son'}, inplace=True)
# df.columns = ['Father', 'Son']
```

```python
df = df * 2.54
df.head()
```

```python
# df = df[df.Gender == 'M'][['Father', 'Height']]
# df = df[['Father', 'Height']]
# df.rename(columns={'Height':'Son'}, inplace=True)

# df.head
```

- 산점도 그리기

```python
plt.scatter(df.Father, df.Son)
plt.grid()
plt.xlabel("Father's height (cm)")
plt.ylabel("Son's height (cm)")
plt.title('Father vs Son', fontsize=15)
plt.show()
```

- 회귀선 구하고 그리기
    - np.linalg.lstsq()
    - np.polyfit()

```python
# y = W * x + b     => 인공지능 기본 형태
weight, bias = np.polyfit(df.Father, df.Son, 1)     # X, Y, 차수
weight, bias
```

```python
# 회귀식을 2차식으로
np.polyfit(df.Father, df.Son, 2)
```

```python
xs = np.array([156, 201])
ys = xs * weight + bias
ys
```

```python
plt.scatter(df.Father, df.Son)
plt.plot(xs, ys, 'r-', lw=3)
plt.grid()
plt.xlabel("Father's height (cm)")
plt.ylabel("Son's height (cm)")
plt.title('Father vs Son', fontsize=15)
plt.show()
```

```python
import seaborn as sns
sns.regplot(x='Father', y='Son', data=df);
```

- Scikit=learn으로 회귀식 구하기

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.get_params()
```

```python
df.shape
```

```python
# 학습 메소드
lr.fit(df[['Father']], df.Son)
```

```python
lr.fit(df.Father.values.reshape(-1, 1), df.Son.values)
```

```python
# coefficient(계수), weight
lr.coef_
```

```python
# intercept(절편), bias
lr.intercept_
```

```python
# 잔차 제곱의 합(RSS: Residual Sum of Square)
lr._residues
```

```python
# 평가 - R squared
lr.score(df.Father.values.reshape(-1, 1), df.Son.values)
```

```python
from sklearn.metrics import r2_score
pred = lr.predict(df.Father.values.reshape(-1, 1))
r2_score(df.Son.values, pred)
```
